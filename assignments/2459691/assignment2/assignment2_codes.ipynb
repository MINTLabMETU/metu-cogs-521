```python
with open ("child_in", "r") as f:
    child = f.read()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

# Create an nlp object
doc = nlp(child)
```

```python
nlp.disable_pipes('tagger', 'parser')
```

```python
nlp.pipe_names
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_)
```
