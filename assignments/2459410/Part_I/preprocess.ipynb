### (a-b) read input from files and store in a dictionary

```python
def read_and_store_input(input_path):
    # initialize dictionary
    data = {}
    # read file names from the input directory
    input_files = glob.glob(input_path+'*.txt')
    for file_path in input_files:
        content = open(file_path, 'r')
        # create an empty list with a key as the name of the current file
        file_name = file_path.split('/')[-1]
        data[file_name]=[]
        # populate the list with content from file
        for line in content:
            # check if line has content
            if line.strip() != "":
                data[file_name].append(line)
    return data

data = read_and_store_input(input_path)
```

## d) Remove punctuation

```python
def tokenize(line):
  #remove punctuation
  line = ''.join(ch for ch in line if ch not in string.punctuation)
  #tokenize
  words = line.split()
  return words

def tokenizer(data):
  #traverse data
  for file, content in data.items():
    for index, line in enumerate(content):
      words = tokenize(line)
      new_line = (" ").join(words)
      content[index] = new_line
```

```python
#call remove function on all data
tokenizer(data)
#see tokenized data after removing punctuation
data 
```

## (e) eliminate trailing white spaces and stop words while deleting dates, numbers

```python
stop_words = []
for line in open(input_path+'stop_words', 'r'):
    # check if not comment
    if line[0] != '#':
        stop_words.append(line.strip())

# traverse the data
for file, content in data.items():
    for index, line in enumerate(content):
        # eliminate trailing white spaces
        line = line.strip() 
        # eliminates dates & numbers
        OutputList = filter(lambda ThisWord: not re.match('^(?:(?:[0-9]{2}[:\/,.]){2}[0-9]{2,4}|am|pm|[0-9]+)$', ThisWord),line.split(" "))
        line = (" ").join(OutputList)
        # eliminate stop words
        new_line = (' ').join(word for word in line.split() if word not in stop_words)
        content[index] = new_line

# lets compare
for index, line in enumerate(data['001.txt']):
    print("LINE #%s: %s" %(index,line))
```
